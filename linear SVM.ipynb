```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
from scipy.io import loadmat
```

# data importing and visulization

```python
data=loadmat('ex6data1.mat')
X=data['X']
y=data['y'].ravel()
sb.scatterplot(x=X[:,0],y=X[:,1],hue=data['y'].ravel())
```

# Make model and prediction

```python
from sklearn.svm import LinearSVC
model=LinearSVC(C=1,max_iter=1000)
model.fit(X,y)
model.score(X,y)
```

```python
#with higher value of C
#it cause overfitting
model1=LinearSVC(C=100,max_iter=100000)
model1.fit(X,y)
model1.score(X,y)
```

```python
temp=model.decision_function(X)
fig,ax=plt.subplots(figsize=(6,5))
ax.scatter(X[:,0],X[:,1],c=temp)
plt.show()
```
